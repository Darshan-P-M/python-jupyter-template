### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'orange']

# Plot
plt.bar(categories, values, color=colors)

# Customization
plt.title('Bar Plot of Category Values')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show plot
plt.show()
```
